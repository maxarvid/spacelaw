## Scraping a list of all national space law available at UNOOSA

Taken from here: http://www.unoosa.org/oosa/en/ourwork/spacelaw/nationalspacelaw/ 

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

%matplotlib inline
pd.set_option('max_colwidth', -1)

from user_agent import generate_user_agent
headers = {'User-Agent': generate_user_agent(device_type="desktop", os=('mac', 'linux'))}
```

```python
url = 'http://www.unoosa.org/oosa/en/ourwork/spacelaw/nationalspacelaw/'

raw_html = requests.get(url, headers=headers)
soup_doc = BeautifulSoup(raw_html.content, 'html.parser')
```

```python
soup_doc.findAll('h3')[0].next_sibling.next
```

```python
national_law_list = [] 

# Iterate through each country with a national space law
for country in soup_doc.findAll('h3')[:-3]:
    print('------------------------')

    # Iterate through each law in said country
    for law in country.next_sibling.next.findAll('strong'):
        law_dict = {}
        if country.string == None:
            law_dict['country'] = 'Austria'
        else:
            law_dict['country'] = country.string
        
        law_dict['name'] = law.text
        
        url_str = law.find_next('a').get('href')
        if url_str[:4] == 'http':
            law_dict['url'] = url_str
        else:
            law_dict['url'] = 'http://www.unoosa.org' + url_str
        #print(law.find_next('a').get('href'))
    
        national_law_list.append(law_dict)
    
        
```
